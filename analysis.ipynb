```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dict = {
    'Sales': [10,20,30,40],
    'Profits': [2,4,3,1],
    'Date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01']
}

df = pd.DataFrame(
    data_dict,
    columns=['Sales', 'Profits'],
    index=data_dict['Date']
)
df
```

```python
fig, ax = plt.subplots(figsize=(5,4))

ax.plot(
    df.index, 
    df['Sales'],
    color='green',
    linewidth=2,
    ls='-.',
    label='Sales'
)
ax.plot(
    df.index, 
    df['Profits'],
    color='red',
    linewidth=2,
    ls='--',
    label='Profits'
)
fig.suptitle(
    'Overall title',
    fontsize=16
)
ax.set_title(
    'Product Sales and Profits',
    fontsize=12
)
ax.legend(['P','S'], loc='upper center')
ax.axvline(1, c='black', ls='dotted')
#ax.text(0.15, 1, 'peak sales -->')
ax.annotate('Peak Sales',
            xy=(1, 3.7),
            xytext=(0.06, 0.7),
            arrowprops=dict(
                facecolor='black',
                width=1,
                headwidth=4,
                connectionstyle='angle3, angleA=300, angleB=1'
            ),
            verticalalignment='center'
           )
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xticks(rotation=45)
plt.show()
```

```python
#twin plot
fig, ax = plt.subplots()

ax.plot(
    df.index, 
    df['Sales'],
    c='green',
    #linewidth=2,
    #ls='-.',
    label='Sales'
)
ax.set_ylabel('Sales')

ax2 = ax.twinx()

ax2.plot(
    df.index, 
    df['Profits'],
    c='red',
    #linewidth=2,
    ls=':',
    label='Profits'
)
ax2.set_ylabel('Profits')

fig.suptitle(
    'Twin plot',
    y=.95,
    fontsize=16
)

fig.legend(bbox_to_anchor=(.68,.5))
plt.show()
```

```python
fig, ax = plt.subplots()
width = .35
x=np.arange(4) #index length
ax.bar(
    x-width/2,
    df['Sales'],
    width=width,
    label='sales',
    alpha=.4 #transparency
)
ax.bar(
    x+width/2,
    df['Profits'],
    width=width,
    label='profits',
    color='yellow'
)
ax.set_title('Stacked Histogram')
ax.set_xticks(x)
ax.set_xticklabels(df.index)
ax.legend()
plt.show() 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cars = pd.read_csv('car_prices.csv',
                   on_bad_lines='skip'
                  )
cars.head(3)
```

```python
cars.info()
```

```python
cars.dtypes
```

```python
cars_numeric = cars.select_dtypes(include=[np.number])
cars_numeric.columns
```

```python
cars_categorical = cars.select_dtypes(include=[object])
cars_categorical.columns
```

```python
correlation = cars_numeric.corr()
```

```python
cars.describe()
```

```python
style_list = ['SUV', 'Sedan', 'Convertible', 'Coupe']
```

```python
#pairplot of relationship btw numeric values
sns.pairplot(
    cars.query('body in @style_list'),
    corner=True,
    hue='body'
)
```

```python
#heatmap of correlation matrix
sns.heatmap(
    #cars.query('body in @style_list'),
    cars_numeric.corr(),
    cmap='YlGnBu',
    annot=True
) 
```

```python
brand_list = ['Ford', 'BMW', 'Toyota', 'Chevrolet']
```

```python
#value retention after mile driven
sns.lmplot(
    x='odometer',
    y='sellingprice',
    data=cars.query('make in @brand_list and sellingprice < 150000 and odometer < 200000').sample(1000),
    hue='make',
    scatter_kws=({'alpha': .3}),
)
```

```python
#checking regression slope
import scipy.stats as stats
cars_regress = cars.dropna()
stats.linregress(x=cars_regress['odometer'], y=cars_regress['sellingprice'])
```

```python
#checking loss per miles driven for each brand 
for brand in brand_list:
    data = cars_regress.loc[cars_regress['make'] == brand]
    print(f"{brand}: {stats.linregress(data['odometer'], data['sellingprice'])[0:2]}")
```

```python
#average price of vehicle by condition on scale from 1 - 5
fig, ax=plt.subplots()

bins = np.linspace(1, 5, 9)
labels = ['1-1.5', '1.5-2', '2-2.5', '2.5-3', '3-3.5', '3.5-4', '4-4.5', '4.5-5']

cars['condition_bins'] = pd.cut(cars['condition'], bins=bins, labels=labels)
sns.barplot(
    data=cars,
    x='condition_bins',
    y='sellingprice'
)
ax.set_title('Average Price by Condition')
sns.despine()
```

```python
top10= cars['make'].value_counts().iloc[:10].index

cars['make'].value_counts()
```

```python
#price by make and condition
avgs = (
        cars.loc[cars['make'].isin(top10)].pivot_table(
            index='make',
            columns='condition_bins',
            values={'sellingprice': 'mean'}
            )
        .droplevel(0, axis=1)
    )
avgs.head()

        
```

```python
fig, ax = plt.subplots()

sns.heatmap(
    avgs,
    annot=True,
    cmap='YlGnBu',
    fmt='g'
)

plt.title('Avg Selling Price by Make and Quality')
```

```python
#ford f-150
f150s = cars[cars['model'] == 'F-150']
f150s.head(3)
```

```python
plt.title('Avg No of Ford-150 sold and price')
sns.histplot(
             data=f150s,
             x='sellingprice',
             color='orange',
             bins=10          
)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot(
            data=f150s,
            x='trim', 
            y='sellingprice'
)

plt.xticks(rotation=90)

plt.title('Ford-150 avg selling price by trims')
plt.show()
```

```python
#trim levels
trim_list = ['XL', 'XLT', 'Platinum', 'SVT Raptor']
```

```python
f150s_list = f150s.query(f"trim in {trim_list}")
```

```python
#Joint Plot of Trim levels
sns.jointplot(
    data=f150s_list,
    x='condition',
    y='sellingprice',
    kind='scatter',
    hue='trim'
)
```

```python
#identifying done deals
sns.set_style('darkgrid')

f150s = f150s.assign(diff_to_mmr = f150s['sellingprice'] - f150s['mmr'])

f150s_by_color = (
    f150s.groupby('color', as_index=False).agg({'diff_to_mmr': 'mean'})
)

fig, ax = plt.subplots()

sns.barplot(
    data=f150s_by_color,
    x='color',
    y='diff_to_mmr'
)

plt.xticks(rotation=90)
plt.show()
```

```python
f150s.head(3)
```

```python
f150s_list = f150s.query("color != 'orange' and trim in @trim_list")

f150s_pivot = (
    f150s_list.loc[f150s_list['year'].between(2009, 2014)].pivot_table(
        index='trim',
        columns='condition_bins',
        values={'sellingprice': 'mean'}
    )
)
```

```python
fig, ax = plt.subplots()
sns.heatmap(
    f150s_pivot,
    annot=True, 
    fmt='g'
)
```

```python
#states in US and how the ford f150 XLT is sold
f150s_XLT= (
    f150s_list.query("trim == 'XLT' and condition >= 3.5")
    .groupby(['state'], as_index=False)
    .agg(mean_diff=('diff_to_mmr', 'mean'),
         auto_count=('diff_to_mmr','count')
    )
)
```

```python
fig, ax=plt.subplots(2)

sns.barplot(
    data=f150s_XLT,
    x='state',
    y='mean_diff',
    ax=ax[0]
)

sns.barplot(
    data=f150s_XLT,
    x='state',
    y='auto_count',
    ax=ax[1]
)
```
